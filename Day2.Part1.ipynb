# Advent of Code

## Day 2: Cube Conundrum

```python
import re

def line_cleaner(original_line):

    cleaned_line = original_line.replace('red','r')
    cleaned_line = cleaned_line.replace('blue','b')
    cleaned_line = cleaned_line.replace('green','g')

    return cleaned_line

with open('Day2.txt') as Day2_file:
    
    game_num = 1
    cubes = [12, 14, 13] # r, b, g
    game_list = []

    for line in Day2_file:

        print(game_num)
        print(cubes)

        num_r = re.findall(r'\s[0-9]+\sr', line)
        print(num_r)
        num_b = re.findall(r'\s[0-9]+\sb', line)
        print(num_b)
        num_g = re.findall(r'\s[0-9]+\sg', line)
        print(num_g)

        r = re.findall(r'[0-9]+', str(num_r))
        print(r)
        b = re.findall(r'[0-9]+', str(num_b))
        print(b)
        g = re.findall(r'[0-9]+', str(num_g))
        print(g)

        r_numbers = [ int(x) for x in r ]
        print(r_numbers)
        b_numbers = [ int(x) for x in b ]
        print(b_numbers)
        g_numbers = [ int(x) for x in g ]
        print(g_numbers)

        r_max = max(r_numbers)
        print(r_max)
        b_max = max(b_numbers)
        print(b_max)
        g_max = max(g_numbers)
        print(g_max)

        if r_max <= cubes[0] and b_max <= cubes[1] and g_max <= cubes[2]:
            game_list.append(game_num)

        print(game_list)

        game_num += 1

        print('\n')

result = sum(game_list)
print(result)
```
