# Advent of Code

## Day 4: Scratchcards

```python
import re

map = ''
seed_to_soil = []
soil_to_fertilizer = []
fertilizer_to_water = []
water_to_light = []
light_to_temperature = []
temperature_to_humidity = []
humidity_to_location = []

with open('Day5.txt') as Day5_file:
    for line in Day5_file:
        if line[:6] == 'seeds:':
            seeds = re.findall(r'[0-9]+', line)
            seeds = [ int(x) for x in seeds]

        if line[:17] == 'seed-to-soil map:':
            map = 'seed-to-soil map:'
        if map == 'seed-to-soil map:':
            seed_to_soil_line = re.findall(r'[0-9]+', line)
            seed_to_soil_line = [ int(x) for x in seed_to_soil_line]
            if seed_to_soil_line != []:
                seed_to_soil.append(seed_to_soil_line)

        if line[:23] == 'soil-to-fertilizer map:':
            map = 'soil-to-fertilizer map:'
        if map == 'soil-to-fertilizer map:':
            soil_to_fertilizer_line = re.findall(r'[0-9]+', line)
            soil_to_fertilizer_line = [ int(x) for x in soil_to_fertilizer_line]
            if soil_to_fertilizer_line != []:
                soil_to_fertilizer.append(soil_to_fertilizer_line)

        if line[:24] == 'fertilizer-to-water map:':
            map = 'fertilizer-to-water map:'
        if map == 'fertilizer-to-water map:':
            fertilizer_to_water_line = re.findall(r'[0-9]+', line)
            fertilizer_to_water_line = [ int(x) for x in fertilizer_to_water_line]
            if fertilizer_to_water_line != []:
                fertilizer_to_water.append(fertilizer_to_water_line)

        if line[:19] == 'water-to-light map:':
            map = 'water-to-light map:'
        if map == 'water-to-light map:':
            water_to_light_line = re.findall(r'[0-9]+', line)
            water_to_light_line = [ int(x) for x in water_to_light_line]
            if water_to_light_line != []:
                water_to_light.append(water_to_light_line)

        if line[:25] == 'light-to-temperature map:':
            map = 'light-to-temperature map:'
        if map == 'light-to-temperature map:':
            light_to_temperature_line = re.findall(r'[0-9]+', line)
            light_to_temperature_line = [ int(x) for x in light_to_temperature_line]
            if light_to_temperature_line != []:
                light_to_temperature.append(light_to_temperature_line)

        if line[:28] == 'temperature-to-humidity map:':
            map = 'temperature-to-humidity map:'
        if map == 'temperature-to-humidity map:':
            temperature_to_humidity_line = re.findall(r'[0-9]+', line)
            temperature_to_humidity_line = [ int(x) for x in temperature_to_humidity_line]
            if temperature_to_humidity_line != []:
                temperature_to_humidity.append(temperature_to_humidity_line)

        if line[:25] == 'humidity-to-location map:':
            map = 'humidity-to-location map:'
        if map == 'humidity-to-location map:':
            humidity_to_location_line = re.findall(r'[0-9]+', line)
            humidity_to_location_line = [ int(x) for x in humidity_to_location_line]
            if humidity_to_location_line != []:
                humidity_to_location.append(humidity_to_location_line)

seed_location = []
for seed in seeds:

    trigger_soil = True
    for soil_seed in seed_to_soil:
        if soil_seed[1] <= seed <= (soil_seed[1] + soil_seed[2]) and trigger_soil:
            soil = soil_seed[0] + seed - soil_seed[1]
            trigger_soil = False
        elif trigger_soil:
            soil = seed

    trigger_fertilizer = True
    for fertilizer_soil in soil_to_fertilizer:
        if fertilizer_soil[1] <= soil <= (fertilizer_soil[1] + fertilizer_soil[2]) and trigger_fertilizer:
            fertilizer = fertilizer_soil[0] + soil - fertilizer_soil[1]
            trigger_fertilizer = False
        elif trigger_fertilizer:
            fertilizer = soil
    
    trigger_water = True
    for water_fertilizer in fertilizer_to_water:
        if water_fertilizer[1] <= fertilizer <= (water_fertilizer[1] + water_fertilizer[2]) and trigger_water:
            water = water_fertilizer[0] + fertilizer - water_fertilizer[1]
            trigger_water = False
        elif trigger_water:
            water = fertilizer
    
    trigger_light = True
    for light_water in water_to_light:
        if light_water[1] <= water <= (light_water[1] + light_water[2]) and trigger_light:
            light = light_water[0] + water - light_water[1]
            trigger_light = False
        elif trigger_light:
            light = water
    
    trigger_temperature = True
    for temperature_light in light_to_temperature:
        if temperature_light[1] <= light <= (temperature_light[1] + temperature_light[2]) and trigger_temperature:
            temperature = temperature_light[0] + light - temperature_light[1]
            trigger_temperature = False
        elif trigger_temperature:
            temperature = light
    
    trigger_humidity = True
    for humidity_temperature in temperature_to_humidity:
        if humidity_temperature[1] <= temperature <= (humidity_temperature[1] + humidity_temperature[2]) and trigger_humidity:
            humidity = humidity_temperature[0] + temperature - humidity_temperature[1]
            trigger_humidity = False
        elif trigger_humidity:
            humidity = temperature
    
    trigger_location = True
    for location_humidity in humidity_to_location:
        if location_humidity[1] <= humidity <= (location_humidity[1] + location_humidity[2]) and trigger_location:
            location = location_humidity[0] + humidity - location_humidity[1]
            trigger_location = False
        elif trigger_location:
            location = humidity

    seed_location.append([seed,location])

seed_location.sort(key=lambda seed_location: seed_location[1])
print(seed_location)
```
